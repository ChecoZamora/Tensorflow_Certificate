#### Visualizing models predictions:
* Take ina  trained model
* Take features and labels
* Create a meshgrid
* of the different x values and make predictions accross the meshgrid
* Plot the predictions as well as a line between each of the zones

```python
# Plotting decision boundary
import numpy as np

def plot_d_b(model, X, y):
  """
  plots the decision boundary created by a model on X
  """
  # Define the axis boundaries of the plot and create a meshgrid
  x_min, x_max = X[:, 0].min() - 0.1, X[:,0].max()+0.1
  y_min, y_max = X[:, 0].min() - 0.1, X[:,0].max()+0.1
  xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                       np.linspace(y_min, y_max, 100))
  # Create the X vale
  x_in = np.c_[xx.ravel(), yy.ravel()]

  # Make predictions
  y_pred = model.predict(x_in)

  # Check for multiclass
  if len(y_pred[0]) > 1:
    print("doing multiclass")
    # Reshape
    y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
  else:
    print("Doing binary")
    y_pred = np.round(y_pred).reshape(xx.shape)

  # Plot
  plt.contourf(xx, yy, y_pred, cmap=plt.cm.RdYlBu, alpha=0.7)
  plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.RdYlBu)
  plt.xlim(xx.min(), xx.max())
  plt.ylim(yy.min(), yy.max())
```

```python
# Check out the preditions
plot_d_b(model_4, X, y)
```

## Using loss curves to find the best learning rate
To find the ideal learning rate, (The rate where the loss decreases the most during training)


**Steps**
1. Callback: Callback is an exgra piece of functionality, you can add to your  *while* its training
2. Another model
3. A modified loss curves plot

```python
# Set random seed
tf.random.set_seed(42)

# Build model
model_q = tf.keras.Sequential([
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")
    ])

# Compile the model

model_q.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["accuracy"])


# Create a learning rate callback
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10**(epoch/20))
# Fit the model
history_9 = model_q.fit(X_train, y_train, epochs=100, callbacks=[lr_scheduler])
```

```python
# Check out the history
pd.DataFrame(history_9.history).plot(xlabel=epochs)
```

```python
pd.DataFrame(history.history).plot()
```

```python
# Plot learning rate vs loss
lrs = 1e-4 * (10**(tf.range(100)/20))
lrs
```

```python
# Plot
plt.figure(figsize=(10, 7))
plt.semilogx(lrs, history_9.history["loss"])
```
